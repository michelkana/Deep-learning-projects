### Galaxy Zoo
For this project we will work with real astronomical data with the purpose of classifying galaxies based on visual attributes as they appear in astronomical surveys of the night sky, such as color and morphology. Galaxy morphology is strongly correlated with many physical parameters of those astronomical bodies, including their mass, density distribution, temperature, and star formation history. In turn, these parameters can be estimated using observations of the attributes mentioned above. 

The Galaxy Zoo project includes collected morphological attributes of nearly 900,000 galaxies drawn from the [Sloan Digital Sky Survey](http://skyserver.sdss.org/dr15/en/home.aspx). The annotation was provided by hundreds of thousands of volunteers, or 'citizen scientists' as we call them, who voted on the galaxy type based only on its image (for more details on galaxy morphology and classification see <a href="https://academic.oup.com/mnras/article/410/1/166/1032478">Chris Lintott et al.</a>). 

Our task is to train a binary classifier on the Galaxy Zoo data in order to distinguish an elliptical galaxy from spirals, mergers and unknown, which we will call non-elliptical galaxies, based on its visual morphology.  

We are provided a subset of data in `galaxy_morphology_labeled.csv` so there is no need for us to download it. The full dataset is available at http://data.galaxyzoo.org.


<span class='sub-q'> Dataset Attributes:</span>
- **green_color**: the amount of light measured in the green frequency band (wavelength range 464 $\pm$ 128~nm)
- **red_color**: the amount of light measured in the red frequency band (wavelength range 658 $\pm$ 138~nm)
- **infrared_color**: the amount of light measured in the infrared frequency band (wavelength range 806 $\pm$ 149~nm)
- **de_vaucouleurs_fit_axial_ratio**: axial ratio associated with the [de Vaucouleurs fit](https://en.wikipedia.org/wiki/De_Vaucouleurs%27_law) to the two-dimensional image of the galaxy
- **exponential_fit_axial_ratio**: the axial ratio associated with an exponential fit to the two-D image of the galaxy 
- **exponential_disc_fit_log_likelihood**: log likelihood associated with an exponential fit to the two-D image of the galaxy
- **de_vaucouleurs_fit_log_likelihood**: log likelihood associated with a de Vaucouleurs fit to the two-D image of the galaxy
- **star_log_likelihood**: the log likelihood that the object is a point-like source (star) object rather than a galaxy
- **number_of_votes**: total number of volunteer votes
- **probability_elliptical**: fraction of votes for elliptical morphology
- **probability_spiral**: fraction of votes for spiral morphology
- **probability_unknown**: fraction of votes for unknown morphology
- **probability_merger**: fraction of votes for merger morphology
- **class**: morphology with the highest fraction of votes

Notice that the volunteers did not have access to any of the attributes but the galaxy shape when voting.

**1.1 Understanding the Problem**

It is a good practice to start by familiarizing theself with the problem in hand. Read the atricle by [Manda Banerji et al.](https://academic.oup.com/mnras/article/406/1/342/1073212) and familiarize ourselves with the attributes in the dataset and their descriptions. Based on the reading, which attributes in the data are most predictive of whether or not a galaxy is elliptical?

Manda Banerji et al. trained a CNN that classifies galaxies into four morphological classes â€“ early types (elliptical), spirals, point sources/artefacts (unknown) and mergers. We are interested in distinguishing elliptical galaxies from the rest.  
 
The de Vaucouleurs profile is commonly used to describe the variation in surface brightness of an elliptical galaxy as a function of radius. The axial ratio `de_vaucouleurs_fit_axial_ratio` and log likelihood `de_vaucouleurs_fit_log_likelihood` associated with the de Vaucouleurs profile will be most predictive for elliptical galaxies.

The exponential profile is used to describe the disc component of a spiral galaxy. The axial ratio `exponential_fit_axial_ratio` and log likelihood `exponential_disc_fit_log_likelihood` associated with the exponential profile will be most predictive for spiral galaxies.

The log likelihood of the object being well fitted by a point spread function helps in distinguishing extended galaxies from more point-like sources. `star_log_likelihood` will be less predictive for clearly distinction of elliptical galaxies. An object that is not point-like might be either elliptical, spiral or merger.

The colors will not be disscriminative without further feature engineering on them.

The number of votes and the fraction of votes per morphology are directly correlated to the response variable. They should not be considered for our classification.

Based on these plots, a naive approach would be to remove all rows where color is outside the range [10 - 20]

```python
def remove_outliers_naive(df, colors=['red_color', 'green_color', 'infrared_color'], col_min = 10, col_max = 20):
    df_copy = df.copy()
    for color in colors:
        df_copy = df_copy[(df_copy[color]>=col_min) & (df_copy[color]<=col_max)]
    print('{} outliers were removed'.format(df.shape[0]-df_copy.shape[0]))
    return df_copy
```

```python
galaxy_df_cleaned = remove_outliers_naive(galaxy_df)
```

The color information seems normally distributed. We can use the Local Outlier Factor method for unsupervized detection.

```python
fig, ax = plt.subplots(1,1)
sns.distplot(galaxy_df['infrared_color'], color="orange", label="Infrared", ax=ax)
sns.distplot(galaxy_df['red_color'], color="red", label="Red", ax=ax)
sns.distplot(galaxy_df['green_color'], color="green", label="Green", ax=ax)
ax.legend()
ax.set_title('Colors distribution');
```

```python
def remove_outliers_lof(df, colors=['red_color', 'green_color', 'infrared_color'], n_neighbors=20, frac_outliers=0.008):
    df_copy = df.copy()
    clf = LocalOutlierFactor(n_neighbors=n_neighbors, contamination=frac_outliers)
    y_pred = clf.fit_predict(df_copy[colors].values)
    y_pred[y_pred==1] = False
    y_pred[y_pred==-1] = True
    df_copy = df_copy[y_pred==False]
    print('{} outliers were removed'.format(df.shape[0]-df_copy.shape[0]))
    return df_copy
```

```python
galaxy_df_cleaned = remove_outliers_lof(galaxy_df)
```

**1.4 Training a Baseline Classifier**

Using `keras`, construct and train a MLP classifier to classify galaxies as either elliptical or non-elliptical based on the expanded set of features.

For the network we want to consider:
- the number of hidden layers
- the number of nodes in each hidden layer
- the activation function (we suggest starting with `relu`)
- the choice of a loss function
- the choice of mini-batch size (we suggest between 1000 and 5000)
- the choice of an optimizer and its parameters, for example: `learning rate`, `momentum`, and `learning rate decay`
- the number of epochs


We split the data into training and test sets, stratifying on the class.

```python
X_train, X_test, y_train, y_test = train_test_split(galaxy_df_cleaned[features], 
                                                    galaxy_df_cleaned[response], 
                                                    test_size=.3, random_state=0,
                                                    stratify = galaxy_df_cleaned[response])
```

We chose to scale the data instead of normalizing. This is helpful when training neural networks, but also for distance calculation when augmenting the data later.

```python
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

We train a MLP classifier to classify galaxies as either elliptical or non-elliptical as follows.

- the number of hidden layers: EDA shows that the data is not linearly separatable. Although 1 layer might be enough, we improve performance by considering **2** hidden layers. This will also support arbitrary decision boundaries.
- the number of nodes in each hidden layer: by rule of thumb, the total number of weights should be less than a small proportion of the model's degree of freedem (ca. 20000 samples x 10 features). We choose **100** for a less complex model, while avoiding underfitting.
- the activation function: **ReLu** shows great performance by cross-validation. Sigmoid and hyperbolic tangent have issues with vanishing gradient, furthermore, sigmoid's gradients are not zero-centered.
- the choice of a loss function: binary crossentropy is the default loss function when doing binary classification. 
- the choice of mini-batch size: at each iteration, a subset of 1000 datapoints is used.
- the choice of an optimizer and its parameters: we start with **stochastic gradient descent**. `learning rate` 1, `momentum` .9 and `nesterov` are found by cross-validation. `learning rate decay` is found as .1/60 epochs. Weights are initialized uniformly.
- the number of epochs: set to **60** for faster performance.

```python
def get_model( 
            lr=.1,
            momentum = .9,
            kernel_initializer='uniform',
            activation=tf.nn.relu,
            nesterov=False
           ):
    mlp = Sequential()
    mlp.add(Dense(100, activation=tf.nn.relu, kernel_initializer=kernel_initializer, input_dim = 10)) 
    mlp.add(Dense(100, kernel_initializer='uniform', activation=activation))
    mlp.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

    mlp.compile(loss='binary_crossentropy',
                  optimizer=SGD(lr=lr, momentum=momentum, decay=lr/60, nesterov=nesterov),
                  metrics=['acc'])  
    return mlp
```

```python
model_CV = KerasClassifier(build_fn=get_model, epochs=60, 
                           batch_size=1000, verbose=1)
kernel_initializer = ['uniform', 'normal', 'zero']
momentum = [.9, .6]
lr = [0.0001, 0.001, 0.01, 0.1, 1, 5]
activation = [tf.nn.relu, tf.nn.sigmoid, tf.nn.tanh]
nesterov = [True, False]
param_grid = dict(kernel_initializer=kernel_initializer, momentum=momentum, lr=lr, activation=activation, nesterov=nesterov)
grid = GridSearchCV(estimator=model_CV, param_grid=param_grid, n_jobs=-1, cv=3)
grid_result = grid.fit(X_train, y_train)
```

```python
print(f'Best Accuracy for {grid_result.best_score_} using {grid_result.best_params_}')
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print(f' mean={mean:.4}, std={stdev:.4} using {param}')
```

The following function fits a MLP with the hyperparameters found via above cross-validation.

```python
def fit_mlp(Xtrain, ytrain, Xtest, ytest, 
            epochs=500,
            batch_size = 1000,
            metrics=['acc'],
            verbose=1,
            optimizer=SGD(lr=1, momentum=.9, decay=1/60, nesterov=False)
           ):
    input_dim = Xtrain.shape[1]

    mlp = Sequential()
    mlp.add(Dense(100, activation=tf.nn.relu, kernel_initializer='uniform', input_dim = input_dim)) 
    mlp.add(Dense(100, kernel_initializer='uniform', activation=tf.nn.relu))
    mlp.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))

    mlp.compile(loss='binary_crossentropy',
                  optimizer=optimizer,
                  metrics=metrics)
    
    mlp_history = mlp.fit(Xtrain, ytrain,
                        batch_size=batch_size,
                        epochs=epochs,
                        verbose=verbose,
                        validation_data=(Xtest, ytest))
    return mlp, mlp_history
```

```python
mlp, mlp_history = fit_mlp(X_train, y_train, X_test, y_test)
```

```python
def plot_mlp(mlp_history):
    fig, ax = plt.subplots(1, 1, figsize=(10,6))
    ax.plot(np.sqrt(mlp_history.history['acc']), 'r', label='train_acc')
    ax.plot(np.sqrt(mlp_history.history['val_acc']), 'b' ,label='val_acc')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('Accuracy')
    ax.legend()
    ax.tick_params()
    ax.set_title('MLP training performance');
    print('Average Accuracy: training {0:0.2%}, validation {1:0.2%}'.format(np.mean(mlp_history.history['acc']), np.mean(mlp_history.history['val_acc'])))
```

```python
plot_mlp(mlp_history)
```

**1.5 Evaluating and Improving The Classifier**

Choose an appropriate metric for evaluating the classifier. Is the average classification accuracy a good metric for this data given the distribution of the classes? Improve the classifier based on the distribution of the classes and on the evaluation of the baseline classifier. 



**Metric**

In classification problems, we can distinguish between the following metrics, whereby the positive class is Elliptical Galaxy and the negative class is Non-Elliptical Galaxy:

   - Recall or Sensitivity or TPR (True Positive Rate): Number of galaxies correctly identified as positive (elliptical) out of total true positives - TP/(TP+FN)

   - Specificity or TNR (True Negative Rate): Number of galaxies correctly identified as negative (non-elliptical) out of total negatives - TN/(TN+FP)

   - Precision: Number of galaxies correctly identified as positive (elliptical) out of total items identified as positive - TP/(TP+FP)

   - False Positive Rate or Type I Error: Number of galaxies wrongly identified as positive (elliptical) out of total true negatives - FP/(FP+TN)

   - False Negative Rate or Type II Error: Number of galaxies wrongly identified as negative (non-elliptical) out of total true positives - FN/(FN+TP)

   - A Confusion Matrix: visual representation of the number of TP, TN, FP and FN.

   - Accuracy: Percentage of total items classified correctly - (TP+TN)/(N+P)

   - F1 Score: Harmonic mean of precision and recall given by - F1 = 2xPrecisionxRecall /(Precision + Recall)

   - ROC-AUC Score: Area under curve of sensitivity (TPR) vs. specificity (TNR).

   - Log-loss: Probabilistic confidence of accuracy. High value of log-loss means that the absolute probabilities have big difference from actual labels.

If we choose the class using absolute probabilistic difference, then we will look at log-loss with care. If the final class prediction is the most important outcome and we donâ€™t want to tune probability threshold, we will rather use AUC score. But if the threshold is well tuned, then F1 will be the scoring to use. In our galaxies classification task, where positive labels (elliptical) are few, we would like our model to predict positive class correctly and hence we will prefer those models which are able to classify these positive labels. 

Accuracy will tend to choose the majority class, because this strategy could provide the overall best classification accuracy. Therefore, **Accuracy is not the most suitable metric for our task**. 

Log-loss usually fails to identify model which produces too many false negatives because the log-loss function is symmetric and does not differentiate between classes. 

Both F1 score and ROC-AUC score can perform well for class imbalance. F1 is better suit for situations where the negative class is small. **We think that ROC-AUC score would make sense as benchmark measure to predict galaxy shape.**

```python
# function adjusts class predictions based on the prediction threshold
def adjust_proba(probs, threshold):
    return [1 if proba >= threshold else 0 for proba in probs]
# function for computing the confusion matrix
def predict_evaluate_cm(model, X, y, threshold=.5):
    classes = ['Non-Elliptical', 'Elliptical']
    y_true = y
    y_pred_proba = model.predict_proba(X)
    y_pred = adjust_proba(y_pred_proba, threshold)
    print("Accuracy on test {0:0.2%}".format(accuracy_score(y_test, y_pred)))
    cm = confusion_matrix(y_true, y_pred)
    cm = cm.astype('float')
    np.set_printoptions(precision=2)
    plt.figure(figsize=(5,5))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Confusion matrix')
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="gray" if cm[i, j] > thresh else "black")

    plt.tight_lawet()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.grid(False)
    plt.rcParams["axes.edgecolor"] = "0.85"
    plt.show()
    
    [fpr, tpr, thr] = roc_curve(y, y_pred_proba)

    idx = np.min(np.where(tpr > threshold))
    plt.figure()
    plt.plot(fpr, tpr, color='coral', label='ROC curve (area = %0.3f)' % auc(fpr, tpr))
    plt.plot([0, 1], [0, 1], 'k--')
    plt.plot([0,fpr[idx]], [tpr[idx],tpr[idx]], 'k--', color='blue')
    plt.plot([fpr[idx],fpr[idx]], [0,tpr[idx]], 'k--', color='blue')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate (1 - specificity)', fontsize=14)
    plt.ylabel('True Positive Rate (recall)', fontsize=14)
    plt.title('Receiver operating characteristic (ROC) curve')
    plt.legend(loc="lower right")
    plt.show()
```

**Evaluation**

Bellow we plot the confusion matrix and AUC ROC metric for our validation set. Our MLP produces 105 false-positives (non-elliptical galaxies predicted as elliptical) and 212 false-negatives (elliptical galaxies predicted as non-elliptical). The ROC-AUC on validation is 98.3%, which is a very good result.

```python
predict_evaluate_cm(mlp, X_test, y_test, threshold=.5)
```

**2.2 Learning rate with SGD optimizer**

```python
fig, ax = plt.subplots(3, 1, figsize=(15,20))
avg_accuracy = []
lrs = [1e1, 1e0, 1e-1, 1e-2, 1e-3, 1e-4]
for lr in lrs:
    sgd = SGD(lr=lr, momentum=0.0, decay=0.0, nesterov=False)
    mlp, mlp_history = fit_mlp(X_train_augmented, y_train_augmented, X_test, y_test, optimizer=sgd, verbose=0, epochs=500)
    ax[0].plot(np.sqrt(mlp_history.history['acc']), label='train_acc - lr {}'.format(lr))
    ax[1].plot(np.sqrt(mlp_history.history['loss']), label='train_loss - lr {}'.format(lr))
    avg_accuracy.append(np.mean(mlp_history.history['acc']))
ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Accuracy')
ax[0].legend()
ax[0].tick_params()
ax[0].set_title('MLP training performance');
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Loss')
ax[1].legend()
ax[1].tick_params()
ax[1].set_title('MLP training performance');
ax[2].scatter(lrs, avg_accuracy)
ax[2].set_xlabel('lr')
ax[2].set_ylabel('Average Accuracy')
ax[2].set_title('MLP training performance - Average Accuracy');
```

**Make an argument for which value of momentum is more appropriate for our data set**

With stochastic gradient descent we compute the derivatives of a small batch of data (1000 galaxies). Momentum improves SGD, especially at those places where the surface curves is more steep for one dimension than for other (ravines at local minima). Momentum can be seen as moving average of the gradients. It adds a direction component, to push the descent towards the minima. Below we try multiple momentum, i.e. a constant factor that is applied to the previous weights before updating the current one. As result we gain faster convergence with less oscillation. Without momentum, the loss reached a plateau around epoch 200 when using learning rate 0.1. With momentum 0.7 we reach the plateau with less than 100 epochs using the same learning rate.

Usually momentum 0.9 is used. In our case, 0.8 works well for the task of classifying elliptical galaxies, providing the best average accuracy on test.

```python
fig, ax = plt.subplots(3, 1, figsize=(15,20))
avg_accuracy = []
mms = [9e-1, 8e-1, 7e-1, 6e-1, 5e-1, 4e-1]
for mm in mms:
    sgd = SGD(lr=0.1, momentum=mm, decay=0.1/500, nesterov=False)
    mlp, mlp_history = fit_mlp(X_train_augmented, y_train_augmented, X_test, y_test, optimizer=sgd, verbose=0, epochs=500)
    ax[0].plot(np.sqrt(mlp_history.history['acc']), label='train_acc - momentum {}'.format(mm))
    ax[1].plot(np.sqrt(mlp_history.history['loss']), label='train_loss - momentum {}'.format(mm))
    avg_accuracy.append(np.mean(mlp_history.history['acc']))
ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Accuracy')
ax[0].legend()
ax[0].tick_params()
ax[0].set_title('MLP training performance');
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Loss')
ax[1].legend()
ax[1].tick_params()
ax[1].set_title('MLP training performance');
ax[2].scatter(mms, avg_accuracy)
ax[2].set_xlabel('momentum')
ax[2].set_ylabel('Average Accuracy')
ax[2].set_title('MLP training performance - Average Accuracy');
```

Below we use both Adam and SGD with a very high amount of epochs and study how the optimizers behave. At epoch 5000, SGD loss is smoother and more stable than Adam's loss. This is because Adam keeps adjusting the learning rate, causing high variations in the gradients, even when the progress towards minima is with very little improvment at the end of trainig. SGD is more stable at the end of training. One could imagine starting training with Adam in order to choose the right direction of the journey and switching to SGD around epoch 3000.

**2.3 Compare model training using Adam versus SGD**

Optimization algorithms' goal is to find the weights which minimize the loss function.

The gradient descent approach starts with a good guess of weight values, it than computes the gradients (partial derivatives of the loss function with regards to those weights). A proportion of the gradients (learning rate) is used to update the weights in the oposite direction of the gradient. We follow the direction of the slope or surface defined by the loss function until we reach the valley (minima).

While batch gradient descent will use the whole training set to calculate the gradients, stochastic gradient descent uses small batches after shuflling the data each time. Weights are updated more frequently then with batch gradient descent, and the loss function can fluctuate heavily. SGD can keep overshooting and not converging. Adding momentum and Nesterov can help addressing issues with SGD.

Adaptive Moment Estimation (Adam) adjusts the learning rate for each weight. The average of past squared gradients and exponentially decaying average of past gradients are used to compute the learning rates. There are three hyperparameters used in this calculation ($\beta_1$,  $\beta_2$ and $\epsilon$) with common values 0.9, 0.999 and 1e-8 respectively.

Usually Adam requires less tuning effort, compared to SGD. Training with Adam is faster, but some [research](https://arxiv.org/abs/1712.07628) suggests that generalization is worse. Another [research](https://arxiv.org/abs/1712.07628) suggests to use both, starting with Adam and switching to SGD when appropriate during training. As explained in this [article](http://ruder.io/optimizing-gradient-descent/) if the data is sparse, Adam (adaptative learning rate) will perform better than SGD. SGD relies on robust initialization and is tricky to tune, however if we care about generalization, SGD might outperform Adam on test. If the model is very complex, training time is critical, and one would prefer the faster Adam.

The following findings can be expressed from our previous plots on the galaxy dataset:

- Adam works well with learning rate 0.001. Other learning rates appear to converge as well, with similar accuracy.
- SGD works best with learning rate 0.1 and momentum 0.9
- SGD has lowest training loss
- Both SGD and Adam converge to a solution after an acceptable number of epochs

We would prefer SGD with learning rate 0.1 and momentum 0.9 as optimizer for the galaxy dataset.

```python
for col in features:
    fig, ax = plt.subplots(1,1)
    sns.distplot(galaxy_df_cleaned[(col)], ax=ax, label='Training', color='blue')
    sns.distplot(universe_df_2[(col)], ax=ax, label='Universe', color='violet') 
    ax.legend()
    ax.set_title(col);
```

The distributions of feature values above suggest that the colors in the universe dataset are from a different distribution than the colors in the training/validation dataset. The remaining features appear to be drawn from the same distribution for both datasets.

This issue is known as dataset shift, especially Covariate (input) shift means that only the colors distribution changes between training and universe dataset, whereas the conditional distribution of the class given the colors remains unchanged.
